```python
import pandas as pd
import gzip
import json

file_path = "C:/Users/Shivangi Dubey/Downloads/receipts.json.gz"  # Replace with the actual file path

# Create an empty DataFrame
df = pd.DataFrame()

# Open the gzipped file
with gzip.open(file_path, 'rb') as f:
    # Read the file content line by line
    for line in f:
        # Decode the line as UTF-8 and remove leading/trailing whitespaces
        decoded_line = line.decode('utf-8').strip()
        
        # Skip empty lines
        if not decoded_line:
            continue
        
        # Load the JSON data from the line
        json_data = json.loads(decoded_line)
        
        # Append the JSON data as a row to the DataFrame
        df = df.append(json_data, ignore_index=True)

# Print the DataFrame
print(df)

print(df.dtypes)

print(df.isnull().sum())

# Check for outliers in numerical columns
print(df.describe())

# Check for unexpected or invalid values
print(df['bonusPointsEarned'].value_counts())
```

```python
df['createDate'] = df['createDate'].apply(lambda x: pd.to_datetime(x['$date'], unit='ms'))
```

```python
df['dateScanned'] = df['dateScanned'].apply(lambda x: pd.to_datetime(x['$date'], unit='ms'))
```

```python
df['modifyDate'] = df['modifyDate'].apply(lambda x: pd.to_datetime(x['$date'], unit='ms'))
```

```python

df['pointsAwardedDate'] = df['pointsAwardedDate'].apply(lambda x: pd.to_datetime(x['$date'], unit='ms') if isinstance(x, dict) else pd.NaT)
```

```python

df['finishedDate'] = df['finishedDate'].apply(lambda x: pd.to_datetime(x['$date'], unit='ms') if isinstance(x, dict) else pd.NaT)
```

```python

df['purchaseDate'] = df['purchaseDate'].apply(lambda x: pd.to_datetime(x['$date'], unit='ms') if isinstance(x, dict) else pd.NaT)
```

```python
# Convert numeric columns to float
numeric_columns = ['bonusPointsEarned', 'pointsEarned', 'purchasedItemCount', 'totalSpent']
df[numeric_columns] = df[numeric_columns].astype(float)

# Handling missing values
df.dropna(subset=['createDate', 'dateScanned', 'totalSpent'], inplace=True)
df['bonusPointsEarnedReason'].fillna('Unknown', inplace=True)

# Standardizing text
df['bonusPointsEarnedReason'] = df['bonusPointsEarnedReason'].str.lower()
df['bonusPointsEarnedReason'] = df['bonusPointsEarnedReason'].str.replace('[^a-zA-Z0-9\s]', '')



# Encoding categorical variables
df['rewardsReceiptStatus'] = df['rewardsReceiptStatus'].astype('category')
df_encoded = pd.get_dummies(df['rewardsReceiptStatus'], prefix='status')
df = pd.concat([df, df_encoded], axis=1)

# Data Validation and Constraints
df = df[df['bonusPointsEarned'] >= 0]
df = df[df['pointsEarned'] >= 0]
df = df[df['purchasedItemCount'] >= 0]

# Data Normalization or Scaling
df['pointsEarned'] = (df['pointsEarned'] - df['pointsEarned'].min()) / (df['pointsEarned'].max() - df['pointsEarned'].min())
df['purchasedItemCount'] = (df['purchasedItemCount'] - df['purchasedItemCount'].mean()) / df['purchasedItemCount'].std()

# Clean rewardsReceiptItemList column
df['rewardsReceiptItemList'] = df['rewardsReceiptItemList'].apply(str)
```

```python
df['_id'] = df['_id'].apply(lambda x: x['$oid'] if isinstance(x, dict) else x)
```
