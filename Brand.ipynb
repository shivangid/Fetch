```python
import pandas as pd
import gzip
import json

file_path = "C:/Users/Shivangi Dubey/Downloads/brands.json.gz" # Replace with the actual file path

# Create an empty DataFrame
df = pd.DataFrame()

# Open the gzipped file
with gzip.open(file_path, 'rb') as f:
    # Read the file content line by line
    for line in f:
        # Decode the line as UTF-8 and remove leading/trailing whitespaces
        decoded_line = line.decode('utf-8').strip()
        
        # Skip empty lines
        if not decoded_line:
            continue
        
        # Load the JSON data from the line
        json_data = json.loads(decoded_line)
        
        # Append the JSON data as a row to the DataFrame
        df = df.append(json_data, ignore_index=True)

# Print the DataFrame
print(df)

print(df.dtypes)

print(df.isnull().sum())

# Check for outliers in numerical columns
print(df.describe())


```

```python
df['_id'] = df['_id'].apply(lambda x: x['$oid'] if isinstance(x, dict) else x)
```

```python
df['cpg'] = df['cpg'].apply(lambda x: x['$id'] if isinstance(x, dict) else x)
```
